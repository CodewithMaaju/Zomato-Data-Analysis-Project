# Step #3  Data Cleaning part1

```python
# Create the user define function
def handleRate(value):
    return round(value)

df['Aggregate rating'] = df['Aggregate rating'].apply(handleRate)
print(df.head())
```

# Question 1 Solution:


Types of Resturant

```python
df.head()
```

```python
#Now making the graph for that we use Seaborn as sns. 
color_mapping = {
    'Dark Green': 'darkgreen',
    'Green': 'green',
    'Yellow': 'yellow',
    'Orange': 'orange',
    'White': 'white',
    'Red': 'red'
}
sns.countplot(x=df['Rating color'], palette=color_mapping)
plt.xlabel('According to the color rating')
plt.show()
```

```python
 # conclusion : Majority of resturant falls in orange color grading
```

# Question no 2

```python
df.head()
```

```python
# Now we will check rating color on xaxis and what is the vote of these 
grouped_data = df.groupby('Rating color')['Votes'].sum()
result = pd.DataFrame({'Votes':grouped_data})
plt.plot(result, c = 'green' , marker = '.')
plt.xlabel('Color rating' , c = 'red' , size = 20)
plt.ylabel('Votes' , c = 'red' , size = 20)
plt.show()
```

```python
# Conclusion: Green color rating resturants have more votes
```

# Question # 5 

```python
# Online and offline 
plt.figure(figsize = (6,5) , )
sns.boxplot(x = 'Has Online delivery' , y = 'Aggregate rating' , data = df)
plt.show()
```

```python
# Conclusion : Online devilver has more good rating. 
```

# Question # 6 

```python
# Rating type with online oder 
#we use heatplot
pivot_table = df.pivot_table(index = 'Aggregate rating' , columns = 'Has Online delivery' , aggfunc= 'size' , fill_value = 0)
sns.heatmap(pivot_table , annot = True , cmap = 'PiYG' , fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Aggregate rating")
plt.ylabel('Has Online delivery')
plt.show()
```

```python
# Conclusion: More offline customer gives 3 rating. 
```
